```python
from premades import train_and_evaluate_lgb, evaluate_strategy_performance
from preprocessing import prep_and_split_data, load_data
from constants import FILENAME

train_df = load_data(FILENAME)
X_train, X_val, y_train, y_val = prep_and_split_data(train_df)

# 1. Define the combinations you want to test
# Rule of thumb for CPU speed: keep n_estimators < 300, max_depth < 8
experiment_params = [
    {
        "n_estimators": 100,
        "learning_rate": 0.1,
        "max_depth": 5,
        "n_jobs": -1,
        "verbose": -1,
    },
    {
        "n_estimators": 200,
        "learning_rate": 0.05,
        "max_depth": 5,
        "n_jobs": -1,
        "verbose": -1,
    },
    {
        "n_estimators": 100,
        "learning_rate": 0.1,
        "max_depth": 3,
        "n_jobs": -1,
        "verbose": -1,
    },  # Shallower trees
]

best_model = None
best_acc = 0

# 2. Run the loop and let your laptop do the work
for params in experiment_params:
    print("-" * 40)
    model, acc = train_and_evaluate_lgb(X_train, y_train, X_val, y_val, params)

    # Optional: Also run your custom finance evaluator here!
    evaluate_strategy_performance(X_val, model.predict(X_val))

    if acc > best_acc:
        best_acc = acc
        best_model = model

print(f"\nBest accuracy found: {best_acc:.4f}")
# Now 'best_model' is saved and ready to predict on test.csv!
```
